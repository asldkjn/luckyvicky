```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 윈도우의 경우
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 데이터 준비
categories = ["가격", "사용감/착용감", "사이즈/무게/개수", "제조/유통/서비스", "편의성/활용성", "품질/디자인/구성", "효과/성능/기능"]

# 긍정 및 부정 절대 수치 예시
dalba_positive = [707, 1414, 192, 9, 1579, 1313, 2814]
biohillbo_positive = [181, 600, 101, 2, 752, 508, 1472]
avene_positive = [935, 2274, 1206, 25, 2878, 1480, 3501]

dalba_negative = [216, 406, 81, 6, 83, 329,333]
biohillbo_negative = [15, 124, 33, 0, 70, 249, 170]
avene_negative = [142, 205, 204, 6, 176, 492, 726]

# 순 긍정 값 계산
dalba_net = [p - n for p, n in zip(dalba_positive, dalba_negative)]
biohillbo_net = [p - n for p, n in zip(biohillbo_positive, biohillbo_negative)]
avene_net = [p - n for p, n in zip(avene_positive, avene_negative)]

# 전체 순 긍정 값 계산
dalba_total = sum(dalba_net)
biohillbo_total = sum(biohillbo_net)
avene_total = sum(avene_net)

# 비율 계산
dalba_ratio = [value / dalba_total for value in dalba_net]
biohillbo_ratio = [value / biohillbo_total for value in biohillbo_net]
avene_ratio = [value / avene_total for value in avene_net]

# 데이터의 첫 값을 끝에 추가하여 원형 연결
dalba_ratio += dalba_ratio[:1]
biohillbo_ratio += biohillbo_ratio[:1]
avene_ratio += avene_ratio[:1]

# 각 축의 각도 계산
num_vars = len(categories)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# 레이더 차트 그리기
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

# 축 레이블 설정
plt.xticks(angles[:-1], categories, color='grey', size=12)

# 비율 데이터 추가
ax.plot(angles, dalba_ratio, linewidth=1, linestyle='solid', color='yellow', label="달바 (비율)")
ax.fill(angles, dalba_ratio, 'yellow', alpha=0.1)

ax.plot(angles, biohillbo_ratio, linewidth=1, linestyle='solid', color='green', label="바이오힐보 (비율)")
ax.fill(angles, biohillbo_ratio, 'green', alpha=0.1)

ax.plot(angles, avene_ratio, linewidth=1, linestyle='solid', color='red', label="아벤느 (비율)")
ax.fill(angles, avene_ratio, 'red', alpha=0.1)

# 타이틀과 범례 추가
plt.title("제품별 카테고리별 상대적 비율 비교", size=15, color='black', y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 윈도우의 경우
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False

# 데이터 준비
categories = ["가격", "사용감/착용감", "사이즈/무게/개수", "제조/유통/서비스", "편의성/활용성", "품질/디자인/구성", "효과/성능/기능"]

# 긍정 및 부정 절대 수치 예시
dalba_positive = [707, 1450, 126, 9, 1579, 1313, 2814]
biohillbo_positive = [181, 632, 57, 2, 752, 504, 1471]
avene_positive = [935, 2326, 1133, 25, 2877, 1480, 3501]

dalba_negative = [216, 421, 51, 6, 83, 329,333]
biohillbo_negative = [15, 131, 17, 0, 70, 249, 170]
avene_negative = [142, 210, 195, 6, 176, 492, 726]

# 순 긍정 값 계산
dalba_net = [p - n for p, n in zip(dalba_positive, dalba_negative)]
biohillbo_net = [p - n for p, n in zip(biohillbo_positive, biohillbo_negative)]
avene_net = [p - n for p, n in zip(avene_positive, avene_negative)]

# 전체 순 긍정 값 계산
dalba_total = sum(dalba_net)
biohillbo_total = sum(biohillbo_net)
avene_total = sum(avene_net)

# 비율 계산 후 로그 변환
small_constant = 1e-3  # 로그 변환 안정성을 위해 작은 상수 추가
dalba_ratio_log = [np.log10(value / dalba_total + small_constant) for value in dalba_net]
biohillbo_ratio_log = [np.log10(value / biohillbo_total + small_constant) for value in biohillbo_net]
avene_ratio_log = [np.log10(value / avene_total + small_constant) for value in avene_net]

# 데이터의 첫 값을 끝에 추가하여 원형 연결
dalba_ratio_log += dalba_ratio_log[:1]
biohillbo_ratio_log += biohillbo_ratio_log[:1]
avene_ratio_log += avene_ratio_log[:1]

# 각 축의 각도 계산
num_vars = len(categories)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# 레이더 차트 그리기
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

# 축 레이블 설정
plt.xticks(angles[:-1], categories, color='grey', size=12)

# 로그 변환된 비율 데이터 추가
ax.plot(angles, dalba_ratio_log, linewidth=1, linestyle='solid', color='yellow', label="달바 (비율, 로그 스케일)")
ax.fill(angles, dalba_ratio_log, 'yellow', alpha=0.1)

ax.plot(angles, biohillbo_ratio_log, linewidth=1, linestyle='solid', color='green', label="바이오힐보 (비율, 로그 스케일)")
ax.fill(angles, biohillbo_ratio_log, 'green', alpha=0.1)

ax.plot(angles, avene_ratio_log, linewidth=1, linestyle='solid', color='red', label="아벤느 (비율, 로그 스케일)")
ax.fill(angles, avene_ratio_log, 'red', alpha=0.1)

# 타이틀과 범례 추가
plt.title("제품별 카테고리별 상대적 비율 비교 (로그 스케일 적용)", size=15, color='black', y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```

```python
import pandas as pd

# 데이터 준비
categories = ["가격", "사용감/착용감", "사이즈/무게/개수", "제조/유통/서비스", "편의성/활용성", "품질/디자인/구성", "효과/성능/기능"]

# 긍정 및 부정 절대 수치 데이터
dalba_positive = [707, 1450, 126, 9, 1579, 1313, 2814]
biohillbo_positive = [181, 632, 57, 2, 752, 504, 1471]
avene_positive = [935, 2326, 1133, 25, 2877, 1480, 3501]

dalba_negative = [216, 421, 51, 6, 83, 329, 333]
biohillbo_negative = [15, 131, 17, 0, 70, 249, 170]
avene_negative = [142, 210, 195, 6, 176, 492, 726]

# 데이터프레임 생성
data = {
    "Category": categories * 3,
    "Brand": ["달바"] * 7 + ["바이오힐보"] * 7 + ["아벤느"] * 7,
    "Positive": dalba_positive + biohillbo_positive + avene_positive,
    "Negative": dalba_negative + biohillbo_negative + avene_negative
}
df = pd.DataFrame(data)

# 점수 계산: 긍정 비율을 기준으로 0~10 사이의 점수로 변환
df['Score'] = ((df['Positive'] / (df['Positive'] + df['Negative'])) * 10).round(2)

# 결과 출력
print("브랜드별 속성별 점수:")
print(df[['Brand', 'Category', 'Score']])
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 윈도우의 경우
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False

# 사용자 선택: 속성 선택 함수 정의
def get_user_selected_attributes():
    print("속성을 선택하세요:")
    for i, category in enumerate(categories):
        print(f"{i + 1}: {category}")
    indices = input("2개의 속성 번호를 콤마로 구분하여 입력하세요 (예: 1, 2): ")
    selected_indices = [int(i) - 1 for i in indices.split(',')]
    return [categories[i] for i in selected_indices]

# 선택된 속성
selected_attributes = get_user_selected_attributes()

# 선택된 속성에 따라 포지셔닝 그래프 그리기
def plot_positioning_map(df, attributes):
    if len(attributes) != 2:
        print("속성은 정확히 2개를 선택해야 합니다.")
        return
    
    # 선택한 속성에 대한 데이터 추출
    attr1, attr2 = attributes
    positioning_data = df[df['Category'].isin(attributes)].pivot(index='Brand', columns='Category', values='Score')
    
    # 포지셔닝 맵 그리기
    plt.figure(figsize=(10, 8))
    for brand in positioning_data.index:
        x = positioning_data.loc[brand, attr1]
        y = positioning_data.loc[brand, attr2]
        plt.scatter(x, y, s=300, label=brand, alpha=0.7)
        plt.text(x, y, brand, ha='center', va='center', fontsize=12, weight='bold')
    
    plt.xlabel(f"{attr1} 점수", fontsize=13)
    plt.ylabel(f"{attr2} 점수", fontsize=13)
    plt.title(f"브랜드 포지셔닝 맵: {attr1} vs {attr2}", fontsize=16)
    plt.legend(title="브랜드")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xlim(0, 10)
    plt.ylim(0, 10)
    plt.show()

# 포지셔닝 맵 그리기
plot_positioning_map(df, selected_attributes)
```

```python
import json
import os
import re
import pandas as pd

folder_path = r"\Users\kyn03\OneDrive\바탕 화면\project_file\interactive_wordcloud"

# 제거할 불용어 리스트 정의
stopwords = ['미스트보다','없어요','막','전에','대신','넘','많아서','없어서','덜']  # 필요에 따라 불용어 추가

# 데이터를 담을 딕셔너리 생성
data = {}
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        path = os.path.join(folder_path, filename)

        with open(path, "r", encoding="utf-8") as file:
            # 파일 이름에서 브랜드, 카테고리, 감정 추출
            match = re.match(r"^([^\_]+)_([\w가-힣_]+)_(positive|negative).*\.json$", filename)
            if match:
                brand = match.group(1)       # 첫 번째 그룹: 브랜드명
                category = match.group(2)    # 두 번째 그룹: 속성명 (다중 언더스코어 포함 가능)
                sentiment = match.group(3)   # 세 번째 그룹: 긍정/부정

                # JSON 파일에서 단어 데이터를 읽어와서 리스트로 저장
                words = json.load(file)
                for word, count in words.items():
                    # 불용어 필터링
                    if word not in stopwords:
                        data.setdefault((brand, category, sentiment), []).append((word, count))
            else:
                print(f"파일명 형식이 맞지 않음: {filename}")

# 데이터를 DataFrame으로 변환
df = pd.DataFrame(columns=["Brand", "Category", "Sentiment", "Word", "Frequency"])

# 데이터프레임에 데이터 추가
rows = []
for (brand, category, sentiment), words in data.items():
    # 빈도수가 높은 순서대로 정렬
    sorted_words = sorted(words, key=lambda x: x[1], reverse=True)
    for word, count in sorted_words:
        rows.append({
            "Brand": brand,
            "Category": category,
            "Sentiment": sentiment,
            "Word": word,
            "Frequency": count
        })

# DataFrame에 데이터를 추가
df = pd.DataFrame(rows)

# Frequency를 숫자형으로 변환
df['Frequency'] = pd.to_numeric(df['Frequency'], errors='coerce')

# 결과 출력
print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from sklearn.feature_extraction.text import TfidfVectorizer
from adjustText import adjust_text
from matplotlib import font_manager, rc
import prince
import unicodedata

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False

# 정규화 함수 정의 (한글 문자 정규화)
def normalize_text(text):
    return unicodedata.normalize('NFC', text)

# Step 1: 데이터 필터링 및 빈도수 기준 필터링
min_frequency = 10
top_n_words = 100  # 각 브랜드별 상위 빈도수 단어 개수로 제한
filtered_df = df[df['Frequency'] >= min_frequency]

# Step 2: TF-IDF 임베딩 생성
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(filtered_df['Word'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Step 3: 브랜드별 상위 빈도수 단어 추출
top_words_df = (
    filtered_df.groupby('Brand', group_keys=False)
    .apply(lambda x: x.nlargest(top_n_words, 'Frequency'))
)

# Step 4: 감성어 중복 여부 체크 및 공통 감성어 필터링
word_counts = top_words_df.groupby('Word')['Brand'].nunique()
common_words = word_counts[word_counts > 1].index
unique_df = top_words_df[~top_words_df['Word'].isin(common_words)]  # 공통 감성어 제거

# Step 5: 브랜드 고유 감성어 피벗 테이블 생성 (이진화 처리)
unique_pivot = unique_df.pivot_table(index='Brand', columns='Word', values='Frequency', aggfunc='sum', fill_value=0)
unique_pivot_binary = unique_pivot.applymap(lambda x: 1 if x > 0 else 0)

# Step 6: MCA 분석 수행
mca = prince.MCA(n_components=2, random_state=42)
mca.fit(unique_pivot_binary)

# 브랜드와 단어 좌표 추출
brand_coords = mca.row_coordinates(unique_pivot_binary)  # 브랜드 좌표
word_coords = mca.column_coordinates(unique_pivot_binary)  # 단어 좌표

# 각 단어 위치에 무작위 값을 추가하여 겹침을 최소화 - 무작위 이동 강도 감소
jitter_strength = 0.01  # 무작위 이동 강도를 매우 작게 설정하여 단어들이 원래 위치에 가깝게 유지되도록 함
word_coords += np.random.normal(0, jitter_strength, word_coords.shape)

# 브랜드별 색상 설정
brand_colors = {
    normalize_text('달바'): '#E1D700',
    normalize_text('바이오힐보'): '#1a535c',
    normalize_text('아벤느'): '#cc7766'
}

# 시각화
plt.figure(figsize=(12, 10))
plt.title("브랜드별 감성어 분포 MCA 분석 그래프", fontsize=16)

# 강제로 축 범위 고정
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

# 브랜드 목록과 좌표 조정 방향 및 크기 설정
brand_adjustments = {
    '달바': {'x_offset': -0.1, 'y_offset': 0},
    '바이오힐보': {'x_offset': -0.2, 'y_offset': -0.3},
    '아벤느': {'x_offset': -0.5, 'y_offset': -0.3}
}

# 각 브랜드의 위치를 조정
for brand, adjustment in brand_adjustments.items():
    if brand in unique_pivot_binary.index:
        brand_index = unique_pivot_binary.index.get_loc(brand)
        brand_coords.iloc[brand_index, 0] += adjustment['x_offset']  # MCA Dimension 1의 좌표 조정
        brand_coords.iloc[brand_index, 1] += adjustment['y_offset']  # MCA Dimension 2의 좌표 조정

# Step 7: 브랜드별 위치 표시 (원형 영역으로 표시)
for i, brand in enumerate(unique_pivot_binary.index):

    center_x, center_y = brand_coords.iloc[i, 0], brand_coords.iloc[i, 1]
    color = brand_colors.get(normalize_text(brand), 'gray')
    
    # 원 생성 (브랜드 색상 적용)
    circle = Circle((center_x, center_y), radius=0.3, edgecolor=color, facecolor='none', linestyle='--', linewidth=1.5)
    plt.gca().add_patch(circle)
    plt.text(center_x, center_y, normalize_text(brand), fontsize=12, ha='center', color=color, weight='bold')

# Step 8: 단어 위치 표시 (부정적 의미에 이탤릭체 적용, 마커 제거 및 adjust_text 사용)
texts = []
# 좌표 수동 조정 예제
for i, word in enumerate(unique_pivot.columns):
    # 특정 단어가 특정 브랜드와 연관이 깊다고 판단되면, 해당 브랜드의 좌표로 이동
    if word in ['오일층이','향이','오일이']:  # 예를 들어 '특정단어1', '특정단어2'가 '아벤느'와 관련이 깊다고 가정
        brand_index = unique_pivot_binary.index.get_loc('달바')
        word_coords.iloc[i, 0] = brand_coords.iloc[brand_index, 0] + np.random.normal(0.1, 0.15)  # 약간의 무작위 이동 추가
        word_coords.iloc[i, 1] = brand_coords.iloc[brand_index, 1] + np.random.normal(0.1, 0.15)
    if word in ['커서', '건조함','대용량','대용량이라','시원하고','온천수라','흡수가','온천수','금방','빠르게','물']:  # 예를 들어 '특정단어1', '특정단어2'가 '아벤느'와 관련이 깊다고 가정
        brand_index = unique_pivot_binary.index.get_loc('아벤느')
        word_coords.iloc[i, 0] = brand_coords.iloc[brand_index, 0] + np.random.normal(0.2, 0.15)  # 약간의 무작위 이동 추가
        word_coords.iloc[i, 1] = brand_coords.iloc[brand_index, 1] + np.random.normal(0.2, 0.15)
    if word in ['보습력이','용량이']:  # 예를 들어 '특정단어1', '특정단어2'가 '아벤느'와 관련이 깊다고 가정
        brand_index = unique_pivot_binary.index.get_loc('바이오힐보')
        word_coords.iloc[i, 0] = brand_coords.iloc[brand_index, 0] + np.random.normal(0.05, 0.1)  # 약간의 무작위 이동 추가
        word_coords.iloc[i, 1] = brand_coords.iloc[brand_index, 1] + np.random.normal(0.05, 0.1)

    # 부정 단어 판별
    is_negative = any(filtered_df[(filtered_df['Word'] == word) & (filtered_df['Sentiment'] == 'negative')].index)
    fontstyle = 'italic' if is_negative else 'normal'
    color = 'dimgray' if not is_negative else 'darkred'  # 부정 단어일 때 색상 변경
    
    # 텍스트 위치에 무작위 좌표 추가
    text = plt.text(word_coords.iloc[i, 0], word_coords.iloc[i, 1], normalize_text(word), 
                    fontsize=10, ha='center', color=color, fontstyle=fontstyle)
    texts.append(text)

# 텍스트 위치 조정 (단어가 그래프 밖으로 나가지 않게 설정)
adjust_text(
    texts,
    only_move={'points': 'xy', 'texts': 'xy'},  # 텍스트 위치 조정 방향을 xy로 설정하여 겹침 방지
    expand_text=(1.5, 1.5),  # 이동 범위를 확대하여 겹침 최소화
    force_text=(0.5, 0.5),  # 텍스트 이동 강도 강화
    lim=1000,  # 텍스트 위치 조정 반복 횟수 최대화
    arrowprops=dict(arrowstyle="-", color='none', lw=0)  # 화살표를 보이지 않게 설정
)

# 축 레이블 및 범례 추가
plt.xlabel("MCA Dimension 1")
plt.ylabel("MCA Dimension 2")
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=normalize_text(brand),
                              markerfacecolor=color, markersize=10) 
                   for brand, color in brand_colors.items()]
plt.legend(handles=legend_elements, title="브랜드", loc="upper right", fontsize=10)
plt.grid(True)
plt.show()




```
