```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

#   
font_path = "C:/Windows/Fonts/malgun.ttf"  #  
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

#   (  )
categories = ["", "/","//", "//","/","//", "//"]
dalba_positive = [76.10, 75.05, 67.61, 60.0, 93.49,77.37,86.27]  #   
biohillbo_positive = [90.05,80.32,70.14,100.0,89.52,64.22,86.28]  #   
avene_positive = [86.33,90.38,84.51,80.65,93.02,70.81,79.17]  #   

dalba_negative = [23.25, 21.55, 28.52, 40.0, 4.91, 19.39, 10.21]  #   
biohillbo_negative = [7.46,16.60,22.92,0.00,8.33,31.48,9.96]  #   
avene_negative = [13.11,8.15,14.30,19.35,5.69,23.54,16.42]  #   


# Calculate net positivity (positive - negative) for each product
dalba_net = [p - n for p, n in zip(dalba_positive, dalba_negative)]
biohillbo_net = [p - n for p, n in zip(biohillbo_positive, biohillbo_negative)]
avene_net = [p - n for p, n in zip(avene_positive, avene_negative)]

# Close the circle by appending the first value to the end of each list
dalba_net += dalba_net[:1]
biohillbo_net += biohillbo_net[:1]
avene_net += avene_net[:1]

# Prepare angles for radar chart
num_vars = len(categories)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Plot the radar chart for each category individually
fig, axs = plt.subplots(3, 3, subplot_kw=dict(polar=True), figsize=(18, 18))
fig.suptitle("     ()", size=20)

for i, (ax, category) in enumerate(zip(axs.flat, categories)):
    # Set category label
    ax.set_title(category, size=15, color='black', y=1.1)
    
    # Single-category net positivity data
    dalba_data = [dalba_net[i]]
    biohillbo_data = [biohillbo_net[i]]
    avene_data = [avene_net[i]]
    
    # Repeat the data to close the radar chart loop
    dalba_data += dalba_data[:1]
    biohillbo_data += biohillbo_data[:1]
    avene_data += avene_data[:1]
    category_angles = [0, 2 * np.pi]  # Only two points to form a circle
    
    # Plot each product's net positivity for this category
    ax.plot(category_angles, dalba_data, linewidth=2, linestyle='solid', color='yellow', label="")
    ax.fill(category_angles, dalba_data, 'yellow', alpha=0.1)
    
    ax.plot(category_angles, biohillbo_data, linewidth=2, linestyle='solid', color='green', label="")
    ax.fill(category_angles, biohillbo_data, 'green', alpha=0.1)
    
    ax.plot(category_angles, avene_data, linewidth=2, linestyle='solid', color='red', label="")
    ax.fill(category_angles, avene_data, 'red', alpha=0.1)

    # Set labels and legend for the first subplot only
    if i == 0:
        ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.2))

# Hide any unused subplots
for j in range(i+1, len(axs.flat)):
    fig.delaxes(axs.flat[j])

plt.show()
```

```python
import json
import os
import re
import pandas as pd

folder_path = r"\Users\kyn03\OneDrive\ \project_file\interactive_wordcloud"

#    
stopwords = ['','','','','','','','','']  #    

#    
data = {}
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        path = os.path.join(folder_path, filename)

        with open(path, "r", encoding="utf-8") as file:
            #   , ,  
            match = re.match(r"^([^\_]+)_([\w-_]+)_(positive|negative).*\.json$", filename)
            if match:
                brand = match.group(1)       #   : 
                category = match.group(2)    #   :  (   )
                sentiment = match.group(3)   #   : /

                # JSON      
                words = json.load(file)
                for word, count in words.items():
                    #  
                    if word not in stopwords:
                        data.setdefault((brand, category, sentiment), []).append((word, count))
            else:
                print(f"   : {filename}")

#  DataFrame 
df = pd.DataFrame(columns=["Brand", "Category", "Sentiment", "Word", "Frequency"])

#   
rows = []
for (brand, category, sentiment), words in data.items():
    #    
    sorted_words = sorted(words, key=lambda x: x[1], reverse=True)
    for word, count in sorted_words:
        rows.append({
            "Brand": brand,
            "Category": category,
            "Sentiment": sentiment,
            "Word": word,
            "Frequency": count
        })

# DataFrame  
df = pd.DataFrame(rows)

# Frequency  
df['Frequency'] = pd.to_numeric(df['Frequency'], errors='coerce')

#  
print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from sklearn.feature_extraction.text import TfidfVectorizer
from adjustText import adjust_text
from matplotlib import font_manager, rc
import prince
import unicodedata

#   
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False

#    (  )
def normalize_text(text):
    return unicodedata.normalize('NFC', text)

# Step 1:      
min_frequency = 10
top_n_words = 100  #       
filtered_df = df[df['Frequency'] >= min_frequency]

# Step 2: TF-IDF  
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(filtered_df['Word'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Step 3:     
top_words_df = (
    filtered_df.groupby('Brand', group_keys=False)
    .apply(lambda x: x.nlargest(top_n_words, 'Frequency'))
)

# Step 4:        
word_counts = top_words_df.groupby('Word')['Brand'].nunique()
common_words = word_counts[word_counts > 1].index
unique_df = top_words_df[~top_words_df['Word'].isin(common_words)]  #   

# Step 5:       ( )
unique_pivot = unique_df.pivot_table(index='Brand', columns='Word', values='Frequency', aggfunc='sum', fill_value=0)
unique_pivot_binary = unique_pivot.applymap(lambda x: 1 if x > 0 else 0)

# Step 6: MCA  
mca = prince.MCA(n_components=2, random_state=42)
mca.fit(unique_pivot_binary)

#    
brand_coords = mca.row_coordinates(unique_pivot_binary)  #  
word_coords = mca.column_coordinates(unique_pivot_binary)  #  

#         -    
jitter_strength = 0.01  #            
word_coords += np.random.normal(0, jitter_strength, word_coords.shape)

#   
brand_colors = {
    normalize_text(''): '#E1D700',
    normalize_text(''): '#1a535c',
    normalize_text(''): '#cc7766'
}

# 
plt.figure(figsize=(12, 10))
plt.title("   MCA  ", fontsize=16)

#    
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

#        
brand_adjustments = {
    '': {'x_offset': -0.1, 'y_offset': 0},
    '': {'x_offset': -0.2, 'y_offset': -0.3},
    '': {'x_offset': -0.5, 'y_offset': -0.3}
}

#    
for brand, adjustment in brand_adjustments.items():
    if brand in unique_pivot_binary.index:
        brand_index = unique_pivot_binary.index.get_loc(brand)
        brand_coords.iloc[brand_index, 0] += adjustment['x_offset']  # MCA Dimension 1  
        brand_coords.iloc[brand_index, 1] += adjustment['y_offset']  # MCA Dimension 2  

# Step 7:    (  )
for i, brand in enumerate(unique_pivot_binary.index):

    center_x, center_y = brand_coords.iloc[i, 0], brand_coords.iloc[i, 1]
    color = brand_colors.get(normalize_text(brand), 'gray')
    
    #   (  )
    circle = Circle((center_x, center_y), radius=0.3, edgecolor=color, facecolor='none', linestyle='--', linewidth=1.5)
    plt.gca().add_patch(circle)
    plt.text(center_x, center_y, normalize_text(brand), fontsize=12, ha='center', color=color, weight='bold')

# Step 8:    (   ,    adjust_text )
texts = []
#    
for i, word in enumerate(unique_pivot.columns):
    #       ,    
    if word in ['','','']:  #   '1', '2' ''   
        brand_index = unique_pivot_binary.index.get_loc('')
        word_coords.iloc[i, 0] = brand_coords.iloc[brand_index, 0] + np.random.normal(0.1, 0.15)  #    
        word_coords.iloc[i, 1] = brand_coords.iloc[brand_index, 1] + np.random.normal(0.1, 0.15)
    if word in ['', '','','','','','','','','','']:  #   '1', '2' ''   
        brand_index = unique_pivot_binary.index.get_loc('')
        word_coords.iloc[i, 0] = brand_coords.iloc[brand_index, 0] + np.random.normal(0.2, 0.15)  #    
        word_coords.iloc[i, 1] = brand_coords.iloc[brand_index, 1] + np.random.normal(0.2, 0.15)
    if word in ['','']:  #   '1', '2' ''   
        brand_index = unique_pivot_binary.index.get_loc('')
        word_coords.iloc[i, 0] = brand_coords.iloc[brand_index, 0] + np.random.normal(0.05, 0.1)  #    
        word_coords.iloc[i, 1] = brand_coords.iloc[brand_index, 1] + np.random.normal(0.05, 0.1)

    #   
    is_negative = any(filtered_df[(filtered_df['Word'] == word) & (filtered_df['Sentiment'] == 'negative')].index)
    fontstyle = 'italic' if is_negative else 'normal'
    color = 'dimgray' if not is_negative else 'darkred'  #     
    
    #     
    text = plt.text(word_coords.iloc[i, 0], word_coords.iloc[i, 1], normalize_text(word), 
                    fontsize=10, ha='center', color=color, fontstyle=fontstyle)
    texts.append(text)

#    (     )
adjust_text(
    texts,
    only_move={'points': 'xy', 'texts': 'xy'},  #     xy   
    expand_text=(1.5, 1.5),  #     
    force_text=(0.5, 0.5),  #    
    lim=1000,  #      
    arrowprops=dict(arrowstyle="-", color='none', lw=0)  #    
)

#     
plt.xlabel("MCA Dimension 1")
plt.ylabel("MCA Dimension 2")
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=normalize_text(brand),
                              markerfacecolor=color, markersize=10) 
                   for brand, color in brand_colors.items()]
plt.legend(handles=legend_elements, title="", loc="upper right", fontsize=10)
plt.grid(True)
plt.show()




```
