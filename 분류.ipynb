```python
import pandas as pd
import os

# 폴더 경로 설정
output_folder = r"\Users\kyn03\OneDrive\바탕 화면\project_file\input"
os.makedirs(output_folder, exist_ok=True)

# 데이터 파일 불러오기
file_path = r"\Users\kyn03\OneDrive\바탕 화면\project_file\달바_결측치처리_최종.xlsx"
df = pd.read_excel(file_path)

# 날짜 데이터를 datetime 형식으로 변환
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y.%m.%d')

# 월과 연도 데이터 추출
df['연도'] = df['날짜'].dt.year
df['월'] = df['날짜'].dt.month

# 1. 월별로 나눈 엑셀 파일 12개 생성
for month in range(1, 13):
    monthly_df = df[df['월'] == month]
    monthly_file = os.path.join(output_folder, f'월별_{month}월_리뷰.xlsx')
    with pd.ExcelWriter(monthly_file) as writer:
        monthly_df.to_excel(writer, index=False)
    print(f"'{monthly_file}' 파일이 생성되었습니다.")

# 2. 봄, 여름, 가을, 겨울로 파일 4개 생성
# 봄(3, 4, 5월), 여름(6, 7, 8월), 가을(9, 10, 11월), 겨울(12, 1, 2월)
season_dict = {
    '봄': [3, 4, 5],
    '여름': [6, 7, 8],
    '가을': [9, 10, 11],
    '겨울': [12, 1, 2]
}

for season, months in season_dict.items():
    season_df = df[df['월'].isin(months)]
    season_file = os.path.join(output_folder, f'{season}_리뷰.xlsx')
    with pd.ExcelWriter(season_file) as writer:
        season_df.to_excel(writer, index=False)
    print(f"'{season_file}' 파일이 생성되었습니다.")

# 3. 연도별로 파일 생성
for year in df['연도'].unique():
    yearly_df = df[df['연도'] == year]
    yearly_file = os.path.join(output_folder, f'{year}_리뷰.xlsx')
    with pd.ExcelWriter(yearly_file) as writer:
        yearly_df.to_excel(writer, index=False)
    print(f"'{yearly_file}' 파일이 생성되었습니다.")
```

```python
import pandas as pd
import os

# 데이터 파일 경로
file_path = r"\Users\kyn03\OneDrive\바탕 화면\project_file\달바_결측치처리_최종.xlsx"
df = pd.read_excel(file_path)

# 저장할 폴더 생성
output_folder = r"\Users\kyn03\OneDrive\바탕 화면\project_file\input"
os.makedirs(output_folder, exist_ok=True)

# 1. 성별로 필터링하여 파일 저장
for gender in ['m', 'f']:
    gender_df = df[df['성별'].notna() & (df['성별'] == gender)]
    gender_file = os.path.join(output_folder, f"성별_{gender}.xlsx")
    gender_df.to_excel(gender_file, index=False)
    print(f"{gender_file} 파일이 저장되었습니다.")

# 2. 연령대별로 필터링하여 파일 저장
for age_group in ['10대', '20대', '30대', '40대 이상']:
    age_df = df[df['연령대'].notna() & (df['연령대'] == age_group)]
    age_file = os.path.join(output_folder, f"연령대_{age_group}.xlsx")
    age_df.to_excel(age_file, index=False)
    print(f"{age_file} 파일이 저장되었습니다.")

# 3. 피부타입별로 필터링하여 파일 저장
for skin_type in ['민감성', '복합성', '건성', '지성']:
    skin_type_df = df[df['피부타입'].notna() & (df['피부타입'] == skin_type)]
    skin_type_file = os.path.join(output_folder, f"피부타입_{skin_type}.xlsx")
    skin_type_df.to_excel(skin_type_file, index=False)
    print(f"{skin_type_file} 파일이 저장되었습니다.")
```
