```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc


# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 윈도우의 경우
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 파일 경로 설정
input_file = r"\Users\kyn03\Downloads\이탈자_리뷰_비공개_제거.xlsx"

# 데이터 읽기
df = pd.read_excel(input_file)

# '피부타입'에서 null 값을 제외하고 데이터 카운트
skin_type_counts = df['피부타입'].dropna().value_counts()

# 파이 차트 그리기
plt.figure(figsize=(8, 8))
plt.pie(
    skin_type_counts,
    labels=skin_type_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Distribution of Skin Types (Non-null values)")
plt.show()
```
